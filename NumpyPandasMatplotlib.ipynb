```python
import pandas as pd  
import numpy as np  
#Creating Series from Array
arr = np.array(['P','a','n','d','a','s'])  
a = pd.Series(arr)  
print("Series from array:")
print(a) 
```

```python
#dictionary
arr = {'x' : 0., 'y' : 2., 'z' : 4.}  
b = pd.Series(arr)  
print("\n\nSeries from dictionary:\n")
print (b)
```

```python
#scalar
x = pd.Series(4, index=[0, 1, 2, 3])  
print("\nSeries using scalar\n")
print (x) 
```

```python
#index 
x = pd.Series([1,2,3],index = ['d','a','c'])  
print("\nSeries through index:") 
print (x)
```

```python
emp = pd.Series(['Parker', 'John', 'Smith', 'William'])   
id =pd.Series([102, 107, 109, 114])     
frame = { 'Emp': emp, 'ID': id }   
result = pd.DataFrame(frame) 
print("\nSeries to Data frame\n")
print(result)
```

```python
 left = pd.DataFrame({  
   'id':[1,2,3,4,5],  
   'Name': ['Ancy', 'Ali', 'Alex', 'Aaron', 'Allen'],  
   'sub':['sub1','sub2','sub4','sub6','sub5']})  
right = pd.DataFrame({  
    'id':[1,2,3,4,5],  
   'Name': ['Brian', 'Ben', 'Bryce', 'Beth', 'Betty'],  
   'sub':['sub2','sub4','sub3','sub6','sub5']})  
print("\n left:\n",left)
print("\n")
print("\n right:\n",right)
print("\n merged:\n",pd.merge(left,right,on='id'))
```
